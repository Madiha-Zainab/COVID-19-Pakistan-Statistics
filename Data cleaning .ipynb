```python
#loading dataset
import pandas as pd
df=pd.read_excel(r'C:\Users\madih\Downloads\DS work\project work\corona-virus-pakistan-dataset-2020\COVID_FINAL_DATA.xlsx')
df.head()
```

```python
#dropping irrelevant column
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head()
```

```python
#renamong the column names
df.columns=['Suspected_Cases','Positive_Cases','Test_Performed','Date','Discharged_Patients','Deaths','Home_Quarantine','New_Suspected_Cases','Region','Admitted_Patients','New_Tests_Performed']
df.head()
```

```python
#checking if there are null values in dataset
print(df.isnull().sum())
```

```python
#filling the null values with zeros
df=df.fillna(0)
print(df.isnull().sum())
```

```python
df
```

```python
#dropping rows with region mobile lab and kptd
df=df[df.Region!='Mobile Lab']
df=df[df.Region!='KPTD']
```

```python
#saving cleaned data
df.to_excel('modified_data.xlsx')
```
